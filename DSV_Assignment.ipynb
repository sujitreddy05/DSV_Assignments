1. Stock Market Analysis
Problem: Analyze stock prices to calculate trends and volatility.
 Tasks:
	Load stock price data (CSV file) using numpy.loadtxt().
	Compute the daily percentage change in stock prices.
	Find moving averages (7-day, 30-day) using numpy.convolve().
	Identify the days with the highest and lowest stock prices.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic dataset of 100 days of stock prices
np.random.seed(42)  # For reproducibility
days = np.arange(1, 101)
initial_price = 100  # Starting stock price
random_changes = np.random.normal(0, 2, size=100)  # Simulating daily price changes
stock_prices = initial_price + np.cumsum(random_changes)  # Cumulative sum to simulate price trend

# Save dataset as a CSV file
np.savetxt("synthetic_stock_prices.csv", stock_prices, delimiter=",")

# Compute the daily percentage change in stock prices
daily_percentage_change = np.diff(stock_prices) / stock_prices[:-1] * 100

# Function to compute moving averages
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Compute 7-day and 30-day moving averages
moving_avg_7 = moving_average(stock_prices, 7)
moving_avg_30 = moving_average(stock_prices, 30)

# Identify the days with the highest and lowest stock prices
highest_price_day = np.argmax(stock_prices)
lowest_price_day = np.argmin(stock_prices)

# Print results
print("Generated Stock Prices:\n", stock_prices)
print("\nDaily Percentage Change:\n", daily_percentage_change)
print("\n7-day Moving Average:\n", moving_avg_7)
print("\n30-day Moving Average:\n", moving_avg_30)
print("\nDay with Highest Price:", highest_price_day + 1, "Price:", stock_prices[highest_price_day])
print("Day with Lowest Price:", lowest_price_day + 1, "Price:", stock_prices[lowest_price_day])

# Plot the stock prices and moving averages
plt.figure(figsize=(10, 5))
plt.plot(days, stock_prices, label="Stock Prices", color="blue")
plt.plot(days[6:], moving_avg_7, label="7-day Moving Avg", color="orange")
plt.plot(days[29:], moving_avg_30, label="30-day Moving Avg", color="red")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.legend()
plt.title("Stock Price Analysis")
plt.show()
```

2. Weather Data Processing
Problem: Analyze temperature variations in a city over the past year.
 Tasks:
	Load the dataset containing daily temperature records.
	Compute the average monthly temperature using NumPy slicing.
	Find the hottest and coldest days of the year.
	Detect temperature outliers using NumPys statistical functions.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic dataset: daily temperatures for one year (365 days)
np.random.seed(42)  # For reproducibility
days = np.arange(1, 366)
mean_temperature = 25  # Average yearly temperature in Celsius
temperature_variation = np.random.normal(0, 5, size=365)  # Random daily variation
temperatures = mean_temperature + np.sin(2 * np.pi * days / 365) * 10 + temperature_variation

# Save dataset as a CSV file
np.savetxt("synthetic_weather_data.csv", temperatures, delimiter=",")

# Compute the average monthly temperature using NumPy slicing
monthly_avg_temp = [np.mean(temperatures[i:i+30]) for i in range(0, 365, 30)]  # Approx. 12 months

# Find the hottest and coldest days of the year
hottest_day = np.argmax(temperatures) + 1  # Adding 1 to match day index
coldest_day = np.argmin(temperatures) + 1

# Detect temperature outliers using Z-score method
z_scores = (temperatures - np.mean(temperatures)) / np.std(temperatures)
outliers = days[np.abs(z_scores) > 2]  # Outliers are more than 2 standard deviations from the mean

# Print results
print("Average Monthly Temperatures:\n", monthly_avg_temp)
print("\nHottest Day:", hottest_day, "Temperature:", temperatures[hottest_day-1])
print("Coldest Day:", coldest_day, "Temperature:", temperatures[coldest_day-1])
print("\nOutlier Days:", outliers)

# Plot temperature trends
plt.figure(figsize=(10, 5))
plt.plot(days, temperatures, label="Daily Temperatures", color="blue")
plt.scatter(outliers, temperatures[outliers - 1], color="red", label="Outliers", marker="o")
plt.xlabel("Days")
plt.ylabel("Temperature (C)")
plt.legend()
plt.title("Temperature Analysis Over a Year")
plt.show()
```

3. Customer Segmentation in E-Commerce
Problem: Identify high-value customers for personalized marketing.
 Tasks:
	Load a dataset containing customer purchase amounts.
	Compute average spending per customer.
	Use NumPy filtering to segment customers into low, medium, and high spenders.
	Find the top 5% of customers with the highest purchases.

```python
import numpy as np

# Generate a synthetic dataset: purchase amounts for 1000 customers
np.random.seed(42)  # For reproducibility
customers = np.arange(1, 1001)  # 1000 customers
purchase_amounts = np.random.exponential(scale=500, size=1000)  # Simulating spending behavior

# Save dataset as a CSV file
np.savetxt("synthetic_customer_data.csv", purchase_amounts, delimiter=",")

# Compute average spending per customer
average_spending = np.mean(purchase_amounts)

# Segment customers into low, medium, and high spenders
low_spenders = customers[purchase_amounts < 200]
medium_spenders = customers[(purchase_amounts >= 200) & (purchase_amounts < 1000)]
high_spenders = customers[purchase_amounts >= 1000]

# Find the top 5% of customers with the highest purchases
threshold = np.percentile(purchase_amounts, 95)  # 95th percentile
top_customers = customers[purchase_amounts >= threshold]

# Print results
print("Average Spending per Customer:", average_spending)
print("\nNumber of Low Spenders:", len(low_spenders))
print("Number of Medium Spenders:", len(medium_spenders))
print("Number of High Spenders:", len(high_spenders))
print("\nTop 5% Customers (High Purchasers):", top_customers)

# Display a histogram of spending behavior
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.hist(purchase_amounts, bins=50, color="blue", alpha=0.7, edgecolor="black")
plt.axvline(threshold, color="red", linestyle="dashed", label="Top 5% Threshold")
plt.xlabel("Purchase Amount ($)")
plt.ylabel("Number of Customers")
plt.legend()
plt.title("Customer Spending Distribution")
plt.show()
```

4. Sports Performance Analysis
Problem: Analyze player performance in a football (soccer) league.
 Tasks:
	Load match data (e.g., goals, assists, passes).
	Compute total goals scored per player using NumPy aggregation.
	Find the most consistent player by calculating standard deviation in goals per match.
	Rank players based on their performance using numpy.argsort().

```python
import numpy as np

# Generate a synthetic dataset: 20 players, 30 matches (random stats)
np.random.seed(42)  # For reproducibility
players = np.array([f"Player_{i+1}" for i in range(20)])  # 20 players
matches = 30  # 30 matches

# Generate random match statistics: goals, assists, passes
goals = np.random.poisson(lam=0.5, size=(20, 30))  # Average 0.5 goals per match
assists = np.random.poisson(lam=0.3, size=(20, 30))  # Average 0.3 assists per match
passes = np.random.randint(20, 80, size=(20, 30))  # 20 to 80 passes per match

# Save dataset as a CSV file
np.savetxt("synthetic_football_data.csv", np.hstack([goals, assists, passes]), delimiter=",")

# Compute total goals scored per player
total_goals = np.sum(goals, axis=1)

# Find the most consistent player (lowest standard deviation in goals per match)
goal_std_dev = np.std(goals, axis=1)
most_consistent_player_index = np.argmin(goal_std_dev)
most_consistent_player = players[most_consistent_player_index]

# Rank players based on total goals
ranked_indices = np.argsort(total_goals)[::-1]  # Sort in descending order
ranked_players = players[ranked_indices]
ranked_goals = total_goals[ranked_indices]

# Print results
print("Total Goals Per Player:\n", dict(zip(players, total_goals)))
print("\nMost Consistent Player:", most_consistent_player, "with Standard Deviation:", goal_std_dev[most_consistent_player_index])
print("\nTop 5 Players Based on Goals:")
for i in range(5):
    print(f"{i+1}. {ranked_players[i]} - {ranked_goals[i]} goals")

# Visualizing the rankings
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.barh(ranked_players[:10], ranked_goals[:10], color="blue")
plt.xlabel("Total Goals")
plt.ylabel("Players")
plt.gca().invert_yaxis()
plt.title("Top 10 Players Based on Goals")
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

# Load image as NumPy array and convert to grayscale
def load_grayscale_image(image_path):
    img = Image.open(image_path).convert("L")  # Convert to grayscale
    img_array = np.array(img, dtype=np.float32)
    return img_array

# Normalize pixel values (0 to 1)
def normalize_image(img_array):
    return img_array / 255.0

# Apply edge detection using Sobel filtering (convolution)
def sobel_edge_detection(img_array):
    sobel_x = np.array([[-1, 0, 1],
                         [-2, 0, 2],
                         [-1, 0, 1]], dtype=np.float32)

    sobel_y = np.array([[-1, -2, -1],
                         [0,  0,  0],
                         [1,  2,  1]], dtype=np.float32)

    edge_x = convolve(img_array, sobel_x)
    edge_y = convolve(img_array, sobel_y)

    edge_magnitude = np.sqrt(edge_x**2 + edge_y**2)
    edge_magnitude = (edge_magnitude / edge_magnitude.max()) * 255  # Normalize for visualization
    return edge_magnitude.astype(np.uint8)

# Resize image using slicing and nearest-neighbor interpolation
def resize_image(img_array, new_size):
    old_size = img_array.shape
    scale_x = old_size[1] / new_size[1]
    scale_y = old_size[0] / new_size[0]

    resized_img = np.zeros(new_size, dtype=np.float32)

    for i in range(new_size[0]):
        for j in range(new_size[1]):
            old_x = int(j * scale_x)
            old_y = int(i * scale_y)
            resized_img[i, j] = img_array[old_y, old_x]

    return resized_img.astype(np.uint8)

# Main execution
image_path = "/content/ducks.jpg"
img_array = load_grayscale_image(image_path)
normalized_img = normalize_image(img_array)
edge_img = sobel_edge_detection(normalized_img)
resized_img = resize_image(img_array, (100, 100))  # Resize to 100x100

# Display results
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(img_array, cmap="gray")
plt.title("Original Grayscale")

plt.subplot(1, 3, 2)
plt.imshow(edge_img, cmap="gray")
plt.title("Sobel Edge Detection")

plt.subplot(1, 3, 3)
plt.imshow(resized_img, cmap="gray")
plt.title("Resized Image (100x100)")

plt.show()
```

Problem: Identify suspicious transactions using statistical analysis.
 Tasks:
	Load transaction data (amount, location, time).
	Compute mean and standard deviation of transaction amounts.
	Identify transactions that are 3 standard deviations above the mean as potential fraud.
	Sort transactions by amount and detect anomalies.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic transaction data (1000 transactions)
np.random.seed(42)  # For reproducibility
transaction_ids = np.arange(1, 1001)
amounts = np.random.normal(loc=100, scale=30, size=1000)  # Mean=100, Std=30
locations = np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami'], size=1000)
times = np.random.randint(0, 24, size=1000)  # Random hour of the day

# Introduce some fraudulent transactions (extreme outliers)
fraud_indices = np.random.choice(1000, 10, replace=False)
amounts[fraud_indices] *= np.random.randint(5, 10)  # Inflate fraud amounts

# Create a DataFrame
df = pd.DataFrame({'Transaction_ID': transaction_ids, 'Amount': amounts, 'Location': locations, 'Time': times})

# Compute mean and standard deviation of transaction amounts
mean_amount = np.mean(df['Amount'])
std_amount = np.std(df['Amount'])

# Identify transactions that are 3 standard deviations above the mean as potential fraud
threshold = mean_amount + (3 * std_amount)
potential_fraud = df[df['Amount'] > threshold]

# Sort transactions by amount to detect anomalies
sorted_transactions = df.sort_values(by='Amount', ascending=False)

# Print summary
print(f"Mean Transaction Amount: {mean_amount:.2f}")
print(f"Standard Deviation: {std_amount:.2f}")
print("\nPotential Fraud Transactions:\n", potential_fraud)

# Plot transactions with fraud threshold
plt.figure(figsize=(10, 5))
plt.hist(df['Amount'], bins=50, color='blue', alpha=0.7, edgecolor='black')
plt.axvline(threshold, color='red', linestyle='dashed', linewidth=2, label="Fraud Threshold")
plt.xlabel("Transaction Amount ($)")
plt.ylabel("Number of Transactions")
plt.legend()
plt.title("Transaction Amount Distribution & Fraud Threshold")
plt.show()
```

7. Sentiment Analysis on Social Media
Problem: Analyze public sentiment using word frequency.
 Tasks:
	Load a dataset of tweets or product reviews.
	Convert text into a word frequency matrix using NumPy.
	Identify the most commonly used words.
	Compute TF-IDF (Term Frequency-Inverse Document Frequency) using NumPy.

```python
import numpy as np
import pandas as pd
import re
from collections import Counter
from sklearn.feature_extraction.text import TfidfTransformer

# Sample dataset: Simulated tweets or product reviews
tweets = [
    "I love this product! It's amazing and works perfectly.",
    "Terrible experience, the product broke after one day!",
    "Highly recommend this to everyone, great quality!",
    "Not worth the price, very disappointing purchase.",
    "Absolutely love it! Will buy again for sure.",
    "Worst purchase ever, I want a refund now!",
    "Decent product, but customer service is horrible.",
    "Great value for money, very satisfied!",
    "Horrible quality, do not buy this!",
    "The product is okay, nothing special."
]

# Preprocess text: Convert to lowercase, remove punctuation and stop words
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    stop_words = set(["this", "is", "it", "and", "to", "for", "the"])  # Basic stop words
    words = [word for word in text.split() if word not in stop_words]
    return words

# Convert text into a word frequency matrix
all_words = []
processed_tweets = [preprocess_text(tweet) for tweet in tweets]
for tweet in processed_tweets:
    all_words.extend(tweet)

word_counts = Counter(all_words)  # Count word occurrences
unique_words = list(word_counts.keys())  # Unique words

# Create a word frequency matrix
word_freq_matrix = np.zeros((len(tweets), len(unique_words)))

for i, tweet in enumerate(processed_tweets):
    for word in tweet:
        if word in unique_words:
            word_freq_matrix[i, unique_words.index(word)] += 1

# Compute TF-IDF using NumPy
tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(word_freq_matrix).toarray()

# Identify most commonly used words
top_words = word_counts.most_common(5)

# Print results
print("Top 5 Most Common Words:", top_words)
print("\nTF-IDF Matrix:\n", tfidf_matrix)

# Convert to DataFrame for better readability
tfidf_df = pd.DataFrame(tfidf_matrix, columns=unique_words)
print("\nTF-IDF DataFrame:\n", tfidf_df.head())
```

8. Transportation Route Optimization
Problem: Optimize delivery routes for an e-commerce company.
 Tasks:
	Load distance matrices for multiple delivery locations.
	Compute the shortest path using NumPys mathematical functions.
	Use broadcasting to calculate time taken for different speed limits.
	Find the best warehouse location by computing average distance to customers.

```python
import numpy as np
import scipy.spatial.distance as dist
from scipy.sparse.csgraph import floyd_warshall

# Sample distance matrix (in km) between 5 delivery locations including a warehouse
# Rows and columns represent locations (0 = warehouse)
distance_matrix = np.array([
    [0, 10, 15, 20, 25],  # Warehouse to others
    [10, 0, 35, 25, 30],  # Location 1
    [15, 35, 0, 30, 20],  # Location 2
    [20, 25, 30, 0, 15],  # Location 3
    [25, 30, 20, 15, 0]   # Location 4
])

# Compute the shortest path using Floyd-Warshall algorithm
shortest_paths = floyd_warshall(distance_matrix, directed=False)

# Define speed limits in km/h
speed_limits = np.array([40, 60, 80])  # Different speed scenarios

# Compute travel times using broadcasting (time = distance / speed)
travel_times = shortest_paths[:, :, np.newaxis] / speed_limits

# Compute the best warehouse location (minimum average distance to customers)
average_distances = np.mean(distance_matrix, axis=1)
best_warehouse_index = np.argmin(average_distances)

# Print results
print("Shortest Paths Matrix:\n", shortest_paths)
print("\nTravel Times for Different Speed Limits (hours):\n", travel_times)
print(f"\nBest Warehouse Location: Location {best_warehouse_index} (Min Avg Distance: {average_distances[best_warehouse_index]:.2f} km)")

```
